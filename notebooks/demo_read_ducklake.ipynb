# Demo: Read Duck Lake parquet from data/ducklake
This notebook shows how to open a DuckDB connection and read a Parquet file placed under data/ducklake.

```python
import duckdb
import os

data_dir = os.environ.get('DATA_DIR', './data')
parquet_path = os.path.join(data_dir, 'ducklake', 'sample.parquet')

print('Looking for:', parquet_path)
# Read into a dataframe (if file exists)
if os.path.exists(parquet_path):
    con = duckdb.connect()
    df = con.execute(f"SELECT * FROM read_parquet('{parquet_path}') LIMIT 10").df()
    display(df)
else:
    print('Parquet not found; drop a sample at', parquet_path)
```
